```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
df.rename(columns={"chronic disease": "chronic_disease", "alcohol consuming": "alcohol_consumption", "shortness of breath": "shortness_of_breath", "swallowing difficulty": "swallowing_difficulty", "chest pain": "chest_pain", "lung_cancer": "target"}, inplace=True)
df
```

```python
df.replace({"gender": {"M": 0, "F": 1}, "target": {"NO": 0, "YES": 1}}, inplace=True)
```

```python
change_2_1_yes_no_to_1_0 = {1: 0, 2: 1}
df.replace({
    "smoking": change_2_1_yes_no_to_1_0,
    "yellow_fingers": change_2_1_yes_no_to_1_0,
    "anxiety": change_2_1_yes_no_to_1_0,
    "peer_pressure": change_2_1_yes_no_to_1_0,
    "chronic_disease": change_2_1_yes_no_to_1_0,
    "fatigue": change_2_1_yes_no_to_1_0,
    "allergy": change_2_1_yes_no_to_1_0,
    "wheezing": change_2_1_yes_no_to_1_0,
    "alcohol_consumption": change_2_1_yes_no_to_1_0,
    "coughing": change_2_1_yes_no_to_1_0,
    "shortness_of_breath": change_2_1_yes_no_to_1_0,
    "swallowing_difficulty": change_2_1_yes_no_to_1_0,
    "chest_pain": change_2_1_yes_no_to_1_0
}, inplace=True)
```

```python
df = df.astype({"gender": "int64", "target": "int64"})
df.info()
```

```python
df['age'] = df['age'] / 100
df['target'].value_counts()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
x = df.loc[:, df.columns != "target"]
y = df["target"]
x_test, x_train, y_test, y_train = train_test_split(x, y, stratify=y, test_size=0.2, random_state=2930)
```

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(kernel="poly", class_weight="balanced", random_state=2930, degree=3))
])
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
grid = GridSearchCV(pipe, param_grid={
    'svc__C': [0.01, 0.1, 1, 10, 100],
    # 'svc__gamma': ['scale', 'auto', 0.01, 0.1, 1, 10],
    'svc__gamma': [1e-4, 1e-3, 1e-2, 1e-1, 1, 1000, 0, "scale"]
}, cv=5, scoring="balanced_accuracy", n_jobs=-1)

grid.fit(x_train, y_train)
display(pd.DataFrame(grid.cv_results_).sort_values(by="mean_test_score", ascending=False))
print("Best parameters:", grid.best_params_)
best_model = grid.best_estimator_
y_pred = best_model.predict(x_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
